```python
# First, install graphviz
!pip install graphviz

# Import the necessary library
from graphviz import Digraph

# Create a new directed graph
dot = Digraph(comment='System Architecture')

# Set the graph to flow from top to bottom
dot.attr(rankdir='TB')

# Add nodes
dot.node('A', 'User')
dot.node('B', 'Frontend SPA')
dot.node('C', 'Backend Server')
dot.node('D', 'NoSQL Database')
dot.node('E', 'LLM API')
dot.node('F', 'File Storage Service')
dot.node('G', 'Data Processing Module')
dot.node('H', 'D3.js Code')
dot.node('I', 'Visualization Engine')

# Add edges
dot.edge('A', 'B', 'Interacts with')
dot.edge('B', 'C', 'Sends requests')
dot.edge('C', 'D', 'Stores data')
dot.edge('C', 'E', 'Integrates with')
dot.edge('C', 'F', 'Stores files')
dot.edge('C', 'G', 'Processes data')
dot.edge('G', 'H', 'Generates')
dot.edge('H', 'B', 'Renders in')
dot.edge('B', 'I', 'Updates')
dot.edge('I', 'A', 'Displays to')

# Render the graph
dot.render('system_architecture', format='png', cleanup=True)

# Display the image
from IPython.display import Image
Image('system_architecture.png')
```
